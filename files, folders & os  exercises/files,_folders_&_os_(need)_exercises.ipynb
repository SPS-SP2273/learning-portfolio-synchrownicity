```python
# TASK 1
# TASK 2
# TASK 3
# TASK 4

import os
import glob
import shutil

month_conversions = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

import os

# Folder and file names
folder_name = 'os-collaboration-exercise-data'
path = os.path.join(folder_name, '*.txt')    # path to os-collaboration-excercise-data
files = glob.glob(path)  # extract all of the files that end in .txt
n = len('os-collaboration-exercise-data')   # Length of the name
for i in range(len(files)):
    old_png_name = files[i]
    if os.path.exists(old_png_name):
    # Rename the PNG file
        new_month = month_conversions[old_png_name[n+1:n+4]]   
        new_png_name = new_month + old_png_name[n+4:]
        new_png_path = os.path.join(folder_name, new_png_name)
        os.rename(old_png_name, new_png_path)
        print(f'Renamed "{old_png_name}" to "{new_png_name}" successfully.')
    
    # Check if the old PNG file exists before renaming
    else:
        print(f'Error: "{old_png_name}" not found in the specified folder.')
```

```python
# TASK 5

countries = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for country in countries:
    path = os.path.join('countries', country)
    try:
        print(f'Creating {path}')
        os.mkdir(path)
    except FileExistsError:
        print(f'{path} already exists; skipping creation.')
```

```python
# TASK 6

all_cairo = glob.glob('os*/*cai*')

for file in all_cairo:
    path_to_destination = os.path.join("countries", "Cairo")
    shutil.copy(file, path_to_destination)
    print(f"Copied file to {path_to_destination}")
```

```python
# TASK 7

# Nested For Loop to do Task 6 for all countries
for country in countries:
    all_country = glob.glob('os*/*' + country[0:3] + '*')
    for file in all_country:
        path_to_destination = os.path.join("countries", country)
        shutil.copy(file, path_to_destination)
        print(f"Copied file to {path_to_destination}")
```
